```python
class A:
  def __init__(self,name,age):
    super().__init__()
    self.name=name
    self.age=age

  def details(self):
    return self.name


class B:
  def __init__(self,name,id):
    super().__init__()
    self.name=name
    self.id=id

  def details(self):
    return self.name

class C(A,B):
  def __init__(self,name,age,id):
    A.__init__(self,name,age)
    B.__init__(self,name,id)

  def get_details(self):
    return self.name,self.id


w=C("anil_ram",25,1)
w.get_details()
```

```python
def area(n):
  x=[]
  for i in range(n):
    y=list(input())
    y.remove(y[1])
    y=[int(i) for i in y]
    x.append(y)
  return x

area(3)
```

```python
def area_merge(n):
  y=area(n)
  for i in y:
    for j in range(n):
      if i[0]<=y[j][0]:
        i[0]=i[0]
        break

    else:
      i[0]=y[j][0]


area_merge(3)
```
